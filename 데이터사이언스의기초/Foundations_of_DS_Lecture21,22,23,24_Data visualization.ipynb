# Scatter plot

- plot: sequence(연속적인 숫자), 1 dimension plot (x)
- scatter plot (x,y): 2 dimension plot (숫자가 연속적이지 않아도 된다.)
- body weight, height 
- See the relationship between numerical variables 

- plot은 숫자가 연속적이어야 하지만, scatter plot은 숫자가 연속적이지 않아도 된다. 
- 예를 들어, body weight, height와 같이...
- 숫자형 변수들간의 관계를 보일 때 사용 

```python
height=[165,177,160,180,185]
weight=[62,67,55,74,90]
```

```python
plt.scatter(height, weight)
plt.xlabel('height(cm)')
plt.ylabel('weight(kg)')
plt.title('Height and weight')
plt.grid(True)
```

```python
## plt.scatter(x,y,s=size_n,c=colors,marker='marker_string',alpha=transparency[0,1](투명도))
```

```python
plt.scatter(height, weight,s=100,c='red',marker='s',alpha=0.5) # 이 형태 유심히 보기 
plt.xlabel('height(cm)')
plt.ylabel('weight(kg)')
plt.title('Height and weight')
plt.grid(True)
```

```python
plt.scatter(height, weight, s=100,c='blue',marker='<',alpha=0.5 )
plt.title("Height and weight")
plt.xlabel('height(cm)')
plt.ylabel('weight(kg)')
plt.grid(True)
```

```python
# Q. plot of city based on the size of population
# (x,y)
```

```python
city=['Seoul','Incheon','Gwangju','Taegu','Pusan',]
lat=[37.56,37.45,35.16,35.87,35.18]
lon=[126.97,126.70,126.85,128.60,129.07]
pop=[16154,2751,2995,128.60,4454] # number of people per km^2 = people/km^2
```

```python
pop_den=np.array(pop)*0.2
```

```python
pop_den
```

zip 함수: zip(lon,lat,city) lon 값 하나, lat 값 하나, city 값 하나 이렇게 세 개의 값을 하나로 묶어주는 함수

```python
colors=['r','g','b','c','k']
plt.scatter(lon,lat,s=pop_den,c=colors,alpha=0.3)
plt.xlabel('Longtitude')
plt.ylabel('Latitude')

# plt.text(x,y,'string')
plt.title('Population density of each city')
for x,y,name in zip(lon,lat,city):
    plt.text(x,y,name)
```

```python
plt.scatter(lon,lat,s=pop_den,c=colors,alpha=0.3) # 원 사이즈를 pop_den 만큼으로 설정 
plt.xlabel('Longtitude')
plt.ylabel('Latitude')
plt.title('Population density of each city')
for x,y,name in zip(lon,lat,city):
    plt.text(x,y,name)
```

## Bar chart

- distribution 을 알기 위함

값의 분포를 알기 위해 사용

```python
import matplotlib.pyplot as plt
```

```python
# ID before_diet after_diet
# M_1, 80, 70
# M_1, 90, 85
# M_1, 70, 67
```

```python
member_IDs=['m_01','m_02','m_03'] # ID
before_diet=[80,90,70]
after_diet=[70,85,67]
```

```python
# plt.bar(x, height)
```

```python
index=np.arange(3)
```

```python
index
```

```python
colors=['b','r','g']
```

```python
plt.bar(index,before_diet,tick_label=member_IDs,color=colors,width=0.5) # width in (0,1)
plt.title('before_diet')
```

```python
plt.bar(index, before_diet, tick_label=member_IDs, color=colors, width=0.6)
```

```python
# compare before_diet with after_diet
```

```python
plt.bar(index,before_diet,tick_label=member_IDs,color='r',width=0.3,align='edge',label='before') 
# width in (0,1), ,align='edge': 각 그래프의 이름의 위치를 두 그래프 사이로 이동시킴
plt.bar(index+0.3,after_diet,tick_label=member_IDs,color='g',width=0.3,align='edge',label='after') 
# width in (0,1)
plt.legend()
plt.xlabel('ID')
plt.ylabel('Weight')
plt.title('Effect of diet')
```

```python
plt.bar(index, before_diet,tick_label=member_IDs, color='r',width=0.3,align='edge',label='before')
plt.bar(index+0.3, after_diet, tick_label=member_IDs, color='g',width=0.3, align='edge', label='before')
plt.title('Effect of diet')
plt.xlabel('ID')
plt.ylabel('Weight')
plt.legend()
```

## Pie chart

```python
fruit=['apple','banana','strawberry','orange']
preference=[7,6,3,2]
```
